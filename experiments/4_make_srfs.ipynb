```python
from mudpy import forward
import numpy as np
import os
```

```python
path = '/Users/sydneydybing/SW4/strain/experiments/'
homo_list = os.listdir(path + 'homo_rupts/srfs/')
rupt_types = []
for file in homo_list:
    if 'slip' in file:
        rupt_types.append(file)
print(rupt_types)
```

```python
# Make homogeneous SRFs

for rtype in rupt_types:
    
    # print(rtype)
    print('-------')
    
    typepath = path + 'homo_rupts/srfs/' + rtype + '/'
    # print(typepath)
    
    rupt_list = os.listdir(typepath)
    print(rupt_list)
    
    for rupt in rupt_list:
        # print(rupt)
        rupt_path = typepath + rupt
        # print(rupt_path)
        log_dir = path + 'homo_rupts/logfiles/'
        rupt_num = rupt.split('_')[2].split('.')[1][-1]
        log_path = log_dir + 'usgs_plane_fine.00000' + str(rupt_num) + '.log'
        # print(log_path)
        
        # Make the SRF
        forward.mudpy2srf(rupt_path, log_path)
    
```

```python
# Make heterogeneous SRFs 

for rtype in rupt_types:
    
    # print(rtype)
    print('-------')
    
    typepath = path + 'hetero_rupts/srfs/' + rtype + '/'
    print(typepath)
    
    rupt_list = os.listdir(typepath)
    # print(rupt_list)
    
    for rupt in rupt_list:
        if 'rupt' in rupt:
            pass
        else:
            continue
        # print(rupt)
        rupt_path = typepath + rupt
        print(rupt_path)
        log_dir = path + 'hetero_rupts/logfiles/'
        rupt_num = rupt.split('_')[2].split('.')[1][-1]
        log_path = log_dir + 'usgs_plane_fine.00000' + str(rupt_num) + '.log'
        print(log_path)
        
        # Make the SRF
        forward.mudpy2srf(rupt_path, log_path)
    
```
