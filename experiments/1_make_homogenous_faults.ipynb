```python
from mudpy import forward
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from obspy.geodetics import degrees2kilometers, gps2dist_azimuth
```

```python
path = '/Users/sydneydybing/SW4/strain/'
out_path = path + 'experiments/'
```

```python
usgs_ff_model = np.genfromtxt(path + 'usgsRC.rupt') # from basic_inversion.param
print(usgs_ff_model)
```

```python
lons = usgs_ff_model[:,1]
lats = usgs_ff_model[:,2]
depths = usgs_ff_model[:,3]

lon_midpoint = min(lons) + (max(lons) - min(lons))/2
lat_midpoint = min(lats) + (max(lats) - min(lats))/2
depth_midpoint = min(depths) + (max(depths) - min(depths))/2
```

```python
plt.scatter(lons, lats)
plt.scatter(lon_midpoint, lat_midpoint)
# plt.scatter(epi_lon, epi_lat)
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.show()

plt.scatter(lons, -depths)
plt.scatter(lon_midpoint, -depth_midpoint)
plt.xlabel('Lon')
plt.ylabel('Depth')
plt.show();
```

```python
Mw = 7.1

# Calculate earthquake rise time using relation in Melgar and Hayes, 2017

# Preferred params for rise time
apref = -5.323
bpref = 0.293

M0 = 10**(1.5*Mw + 9.05)
# print(M0)

logS = apref + bpref * np.log10(M0)

rise_time = 10**logS

print(rise_time)    
```

```python
# Get edges of fault

min_lon = min(lons)
min_lat = min(lats)
min_depth = min(depths)

max_lon = max(lons)
max_lat = max(lats)
max_depth = max(depths)

# print(min_lon, max_lon)
# print(min_lat, max_lat)
# print(min_depth, max_depth)

# Fault length along strike
fault_len = gps2dist_azimuth(min_lat, min_lon, max_lat, max_lon)[0] / 1000
print(fault_len)

# Fault depth extent
fault_depth = max_depth - min_depth
print(fault_depth)

# Figure out how many along-strike subfaults for your chosen spacing
dx_strike = 1 # km
nstrike = int(fault_len/dx_strike)
print(nstrike)

# Figure out how many along-dip subfaults for your chosen spacing
dx_dip = 0.5
ndip = int(fault_depth/dx_dip)
print(ndip)
num_updip = int(ndip / 2)
print(num_updip)
num_downdip = num_updip
```

```python
# Make fault

fout = out_path + 'usgs_plane_fine.fault'

forward.makefault(fout,strike,dip,nstrike,dx_dip,dx_strike,epicenter,num_updip,num_downdip,rise_time)
```

```python
# Check how it looks

new_fault = np.genfromtxt(fout)

new_lons = new_fault[:,1]
new_lats = new_fault[:,2]
new_depths = new_fault[:,3]

plt.scatter(new_lons, new_lats, s = 0.5)
plt.scatter(lon_midpoint, lat_midpoint)
# plt.scatter(epi_lon, epi_lat)
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.show()

plt.scatter(new_lons, -new_depths, s = 0.5)
plt.scatter(lon_midpoint, -depth_midpoint)
plt.xlabel('Lon')
plt.ylabel('Depth')
plt.show();
```

```python
# Find the subfault closest to the desired hypocenter for force_hypocenter = True

# Find the hypocenter from the coarse rupt model

# .rupt columns:
# No,lon,lat,z(km),strike,dip,rise,dura,ss-slip(m),ds-slip(m),ss_len(m),ds_len(m),rupt_time(s),rigidity(Pa)

onset_times = usgs_ff_model[:,12]
# print(onset_times)

# Next need to find the hypocenter of this fault segment - look for where the lowest rupt_time(s) is (rupture start time)
earliest_rupt_start_time = min(onset_times)
m = np.where(onset_times == earliest_rupt_start_time)[0][0]

# print(usgs_ff_model[m])

hypo_lon = usgs_ff_model[m][1]
hypo_lat = usgs_ff_model[m][2]
hypo_depth = usgs_ff_model[m][3] # km

print(hypo_lon, hypo_lat, hypo_depth)

# Find closest subfault in new fine fault to that subfault

hypo = np.array([hypo_lon, hypo_lat, hypo_depth])

dists = []

for idx in range(len(new_fault)):
    
    sub_lon = new_fault[idx,1]
    sub_lat = new_fault[idx,2]
    sub_depth = new_fault[idx,3]
    
    subfault = np.array([sub_lon, sub_lat, sub_depth])
    
    dist = np.linalg.norm(hypo - subfault)
    
    dists.append(dist)

j = np.where(np.array(dists) == min(np.array(dists)))[0]
print(j)

print('Hypocenter subfault coordinates for force_hypocenter:')
print(new_fault[j,1], new_fault[j,2], new_fault[j,3])
    
plt.scatter(new_lons, new_lats, s = 0.5, label = 'All subfaults')
plt.scatter(lon_midpoint, lat_midpoint, s = 8, label = 'Midpoint')
plt.scatter(hypo_lon, hypo_lat, s = 8, label = 'Actual hypocenter')
plt.scatter(new_fault[j,1], new_fault[j,2], s = 8, label = 'Closest subfault to hypocenter')
# plt.scatter(epi_lon, epi_lat)
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.legend()
plt.show()

plt.scatter(new_lons, -new_depths, s = 0.5, label = 'All subfaults')
plt.scatter(lon_midpoint, -depth_midpoint, s = 8, label = 'Midpoint')
plt.scatter(hypo_lon, -hypo_depth, s = 8, label = 'Actual hypocenter')
plt.scatter(new_fault[j,1], -new_fault[j,3], s = 8, label = 'Closest subfault to hypocenter')
plt.xlabel('Lon')
plt.ylabel('Depth')
plt.legend()
plt.show();
```
