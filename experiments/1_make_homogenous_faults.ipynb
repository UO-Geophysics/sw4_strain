```python
from mudpy import forward
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from obspy.geodetics import degrees2kilometers, gps2dist_azimuth
```

```python
path = '/Users/sydneydybing/SW4/strain/'
out_path = path + 'experiments/'
```

```python
usgs_ff_model = np.genfromtxt(path + 'usgsRC.rupt') # from basic_inversion.param
print(usgs_ff_model)
```

```python
lons = usgs_ff_model[:,1]
lats = usgs_ff_model[:,2]
depths = usgs_ff_model[:,3]

lon_midpoint = min(lons) + (max(lons) - min(lons))/2
lat_midpoint = min(lats) + (max(lats) - min(lats))/2
depth_midpoint = min(depths) + (max(depths) - min(depths))/2
```

```python
plt.scatter(lons, lats)
plt.scatter(lon_midpoint, lat_midpoint)
# plt.scatter(epi_lon, epi_lat)
plt.xlabel('Lon')
plt.ylabel('Lat')
plt.show()

plt.scatter(lons, -depths)
plt.scatter(lon_midpoint, -depth_midpoint)
plt.xlabel('Lon')
plt.ylabel('Depth')
plt.show();
```

```python
Mw = 7.1

# Calculate earthquake rise time using relation in Melgar and Hayes, 2017

# Preferred params for rise time
apref = -5.323
bpref = 0.293

M0 = 10**(1.5*Mw + 9.05)
# print(M0)

logS = apref + bpref * np.log10(M0)

rise_time = 10**logS

print(rise_time)    
```

```python
# Get edges of fault

min_lon = min(lons)
min_lat = min(lats)
min_depth = min(depths)

max_lon = max(lons)
max_lat = max(lats)
max_depth = max(depths)

# print(min_lon, max_lon)
# print(min_lat, max_lat)
# print(min_depth, max_depth)

# Fault length along strike
fault_len = gps2dist_azimuth(min_lat, min_lon, max_lat, max_lon)[0] / 1000
print(fault_len)

# Fault depth extent
fault_depth = max_depth - min_depth
print(fault_depth)

# Figure out how many along-strike subfaults for your chosen spacing
dx_strike = 1 # km
nstrike = int(fault_len/dx_strike)
print(nstrike)

# Figure out how many along-dip subfaults for your chosen spacing
dx_dip = 0.5
ndip = int(fault_depth/dx_dip)
print(ndip)
num_updip = int(ndip / 2)
print(num_updip)
num_downdip = num_updip
```

```python
# Make fault

fout = out_path + 'usgs_plane_fine.fault'

forward.makefault(fout,strike,dip,nstrike,dx_dip,dx_strike,epicenter,num_updip,num_downdip,rise_time)
```
