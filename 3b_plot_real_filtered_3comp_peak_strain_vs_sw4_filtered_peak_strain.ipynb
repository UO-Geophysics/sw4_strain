```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from obspy.core import read, Stream
```

```python
sw4_path = '/Users/sydneydybing/SW4/strain/'
real_path = '/Users/sydneydybing/StrainProject/2024/'
```

```python
# Stations

sta_file = 'sw4_strain_stations_metadata.csv'
stas = pd.read_csv(sw4_path + sta_file)
stas
```

```python
# Subplot version

count = 0
fig = plt.figure(figsize = (8,8), dpi = 100, facecolor = 'white')

for idx in range(len(stas)):
    
    sta = stas.BNUM.values[idx]
    theo_p_arr = stas.p_arrival.values[idx] # Seconds after origin time of earthquake
    hypo_dist = stas.hypo_dist_km.values[idx]
    
    # Noisy filtered SW4 data
    sw4_rms_strain = np.load(sw4_path + 'final_RC_model_outputs_processed/strain/noise_added/filtered/rms/' + str(sta) + '.npy')
    sw4_peak_strain = np.load(sw4_path + 'final_RC_model_outputs_processed/strain/noise_added/filtered/peak/' + str(sta) + '.npy')
    
    try:
        real_rms_strain = read(real_path + 'strain_data/3comp_rms/2019-07-06_M7.1/' + sta + '.mseed')
        real_peak_strain = read(real_path + 'strain_data/3comp_peak_post_p/2019-07-06_M7.1/' + sta + '.mseed')
        # real_rms_strain.plot()
        # real_peak_strain.plot()
    except:
        print('No real data for station ' + sta)
        continue
    
    count += 1
    
    # Load the SW4 GNSS data for the strain station locations
    
    sw4_rms_gnss = np.load(sw4_path + 'final_RC_model_outputs_processed/gnss/rms/' + str(sta) + '.npy')
    sw4_peak_gnss = np.load(sw4_path + 'final_RC_model_outputs_processed/gnss/peak/' + str(sta) + '.npy')
    
    sw4_rms_gnss_times = sw4_rms_gnss[:,0]
    sw4_rms_gnss_data = sw4_rms_gnss[:,1]
    
    sw4_peak_gnss_times = sw4_peak_gnss[:,0]
    sw4_peak_gnss_data = sw4_peak_gnss[:,1]
    
    # RMS data both start at the earthquake origin time
    
    sw4_rms_strain_times = sw4_rms_strain[:,0]
    sw4_rms_strain_data = sw4_rms_strain[:,1]
    
    real_rms_strain_times = real_rms_strain[0].times()
    real_rms_strain_data = real_rms_strain[0].data
    
    sw4_peak_strain_times = sw4_peak_strain[:,0]
    sw4_peak_strain_data = sw4_peak_strain[:,1]
    
    real_peak_strain_times = real_peak_strain[0].times() # This has been shifted so that the P-wave arrival is at 10 seconds
    real_peak_strain_data = real_peak_strain[0].data
    
    # Shift the SW4 peak strain so that it lines up with the real peak strain, which has the P-wave arrival at 10 seconds
    
    if theo_p_arr >= 10: # Shift the peak strain backward
        shift_amount = theo_p_arr - 10
        sw4_rms_strain_times = sw4_rms_strain_times - shift_amount
        sw4_peak_strain_times = sw4_peak_strain_times - shift_amount
        sw4_rms_gnss_times = sw4_rms_gnss_times - shift_amount
        sw4_peak_gnss_times = sw4_peak_gnss_times - shift_amount
    else: # Shift the peak strain forward
        shift_amount = 10 - theo_p_arr
        sw4_rms_strain_times = sw4_rms_strain_times + shift_amount
        sw4_peak_strain_times = sw4_peak_strain_times + shift_amount
        sw4_rms_gnss_times = sw4_rms_gnss_times + shift_amount
        sw4_peak_gnss_times = sw4_peak_gnss_times + shift_amount
    
    # Normalize data option
    norm_sw4_peak_data = np.copy(sw4_peak_strain_data)
    norm_sw4_peak_level = sw4_peak_strain_data[0]
    norm_sw4_peak_data = norm_sw4_peak_data - norm_sw4_peak_level + 1*10**-9
    
    norm_sw4_peak_gnss_data = np.copy(sw4_peak_gnss_data)
    norm_sw4_peak_gnss_level = sw4_peak_gnss_data[0]
    norm_sw4_peak_gnss_data = norm_sw4_peak_gnss_data - norm_sw4_peak_gnss_level + 1*10**-9
    
    norm_real_peak_data = np.copy(real_peak_strain_data)
    norm_real_peak_level = real_peak_strain_data[0]
    norm_real_peak_data = norm_real_peak_data - norm_real_peak_level + 1*10**-9
    
    plot_norm = False
    plot_peak = True

    ax = fig.add_subplot(3,2,count)
    plt.grid(alpha = 0.5, zorder = 0)
    
    if plot_peak:
        if plot_norm:
            ax.plot(real_peak_strain_times, norm_real_peak_data, color = 'red', label = 'Norm real peak strain')
            ax.plot(sw4_peak_strain_times, norm_sw4_peak_data, color = 'blue', label = 'Norm SW4 peak strain')
            ax.plot(sw4_peak_gnss_times, norm_sw4_peak_gnss_data, color = 'green', label = 'Norm SW4 peak displacement')
        else:
            ax.plot(real_peak_strain_times, real_peak_strain_data, color = 'red', label = 'Real peak strain')
            ax.plot(sw4_peak_strain_times, sw4_peak_strain_data, color = 'blue', label = 'SW4 peak strain')
            # ax2 = ax.twinx()
            # ax2.plot(sw4_peak_gnss_times, sw4_peak_gnss_data, color = 'green', label = 'SW4 peak displacement')
        ax.semilogy()
        # ax2.semilogy()
    else:
        ax.plot(real_rms_strain_times, real_rms_strain_data, color = 'red', label = 'Real RMS strain')
        ax.plot(sw4_rms_strain_times, sw4_rms_strain_data, color = 'blue', label = 'SW4 RMS strain')
        # ax2 = ax.twinx()
        # ax2.plot(sw4_rms_gnss_times, sw4_rms_gnss_data, color = 'green', label = 'SW4 RMS\ndisplacement')
    
    ax.axvline(10, color = 'gray', linestyle = '--', label = 'P-wave arrival')
    ax.set_xlim(5, 40)
    ax.text(x = 0.65, y = 0.1, s = sta, fontsize = 20, transform = ax.transAxes) 
    
    if count == 1:
        ax.text(x = -0.32, y = 0.9, s = '(a)', fontsize = 20, transform = ax.transAxes)
    elif count == 2:
        ax.text(x = -0.32, y = 0.9, s = '(b)', fontsize = 20, transform = ax.transAxes)
        if plot_norm:
            ax.set_ylabel('Relative peak strain', fontsize = 14)
            # ax2.set_ylabel('Relative peak displacement (m)', fontsize = 14, rotation = 270)
        else:
            ax.set_ylabel('Peak strain', fontsize = 14)
            # ax2.set_ylabel('Peak displacement (m)', fontsize = 14, rotation = 270)
    elif count == 3:
        ax.text(x = -0.32, y = 0.9, s = '(c)', fontsize = 20, transform = ax.transAxes)
        if plot_norm:
            ax.set_ylabel('Relative peak strain', fontsize = 14)
            # ax2.set_ylabel('Relative peak displacement (m)', fontsize = 14, rotation = 270)
        else:
            ax.set_ylabel('Peak strain', fontsize = 14)
            # ax2.set_ylabel('Peak displacement (m)', fontsize = 14, rotation = 270)
        
        ax.legend(bbox_to_anchor = [2.2, -0.3])
        # ax2.legend(bbox_to_anchor = [2.2, -0.7])
        
    elif count == 4:
        ax.text(x = -0.32, y = 0.9, s = '(d)', fontsize = 20, transform = ax.transAxes)
    elif count == 5:
        ax.text(x = -0.32, y = 0.9, s = '(e)', fontsize = 20, transform = ax.transAxes)
    
    if count == 1 or count == 2 or count == 3:
        ax.tick_params(axis = 'both', bottom = False, labelbottom = False)
    elif count == 4 or count == 5:
        ax.set_xlabel('Time (s)', fontsize = 14)
    
plt.subplots_adjust(hspace = 0, wspace = 0.75)

plt.show()
```
